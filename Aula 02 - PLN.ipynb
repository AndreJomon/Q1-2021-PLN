#Tarefa 1
- Buscar todas as ocorrências de palavras terminadas em “mento“, "agem" e "ção".
Qual ocorre mais?


```
textBrega = "O juramento de um Deus é a exceção de todos nós. A carruagem celeste virá descendo por terra em busca de ti. Nesse momento a ferrugem de sua alma testará a aprovação de coragem e envenenamento. Moçãomento"

wordList0 = re.findall("[-'a-zA-ZÀ-ÖØ-öø-ÿ0-9]+ento\\b", FileContent)
wordList1 = re.findall("[-'a-zA-ZÀ-ÖØ-öø-ÿ0-9]+agem\\b", FileContent)
wordList2 = re.findall("[-'a-zA-ZÀ-ÖØ-öø-ÿ0-9]+ção\\b", FileContent)
```

```
print("O número de palavras que terminam em 'ento' é: {}".format(len(wordList0)))
print("O número de palavras que terminam em 'agem' é: {}".format(len(wordList1)))
print("O número de palavras que terminam em 'ção' é: {}".format(len(wordList2)))
```
